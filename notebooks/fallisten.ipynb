```python
#pruefliste["Fallnummer"] = ki["Fallnummer"]
#pruefliste["HDL_Fallnr"] = ki["HDL NR"]
from collections import defaultdict
d = defaultdict(list)
ld_dict = defaultdict(lambda x: False)
```

```python
for row in ki.iterrows():
    fallnr = row[1]["Fallnummer"]
    grund = row[1]["SIM-Grund"]
    d[fallnr].append(grund)
    
```

```python
for row in p21.iterrows():
    fallnr = row[1]["Fallnummer"]
    grund = row[1]["SIM-Grund"]
    d[fallnr].append(grund)
```

```python
for row in lab_txt.iterrows():
    fallnr = row[1]["Fallnummer"]
    grund = row[1]["SIM-Grund"]
    d[fallnr].append(grund)
    ld_dict[fallnr]=True
```

```python
dd = pd.DataFrame.from_dict(d, orient="index")
```

```python
pruefliste["Fallnummer"] = dd.index
pruefliste["No 01"] = dd[0].values
pruefliste["No 02"] = dd[1].values
pruefliste["No 03"] = dd[2].values
pruefliste["No 04"] = dd[3].values
pruefliste["No 05"] = dd[4].values


```

```python
pruefliste.to_csv("pruefliste.csv", encoding="latin1", sep=";", index=False)
```
