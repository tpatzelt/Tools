```python
lab = pd.read_csv("labor/Labor_2018_mit_Pruefgrund.csv",encoding="latin1",sep=";",dtype=str)

lab.dropna(subset=["Prüfgrund"],inplace=True)
lab["Info zu Prüfgrund"] = lab["Laborkürzel"].astype(str) + ": " + lab["Info zu Prüfgrund"]

lab
```

```python
lab["Prüfgrund"].loc[lab["Prüfgrund"]=="N17.91, N17.92, N17.93"] = "N17.93"
lab["Prüfgrund"].loc[lab["Prüfgrund"]=="N17.91, N17.92"] = "N17.92"
lab["Prüfgrund"].loc[lab["Prüfgrund"]=="N17.91, N17.91, N17.92"] = "N17.91"
```

```python
lab[lab["Prüfgrund"]=="N17.91, N17.91, N17.92, N19.93"]
```

```python
lab.rename(columns={"KH-ID":"Fallnummer", "Info zu Prüfgrund":"Hinweis (Regelbezeichnung)"},index=str,inplace=True)
lab =lab.groupby(['Fallnummer',"Prüfgrund"])["Hinweis (Regelbezeichnung)"].apply(', '.join).reset_index()
lab["Prüfgrund (Regel-Nr.)"] = lab["Prüfgrund"].apply(lambda x: "SIM_LAB_" + str(x))
print(lab.columns)
```

```python
j = 0
for index,row in pruefliste.loc[pruefliste["Regelherkunft"].isin(["LAB"])].iterrows():
    info = lab["Hinweis (Regelbezeichnung)"].loc[(lab["Prüfgrund (Regel-Nr.)"] == row["Prüfgrund (Regel-Nr.)"]) & (lab["Fallnummer"] == row["Fallnummer"])]
    if len(info)<1:
        continue
    row["Hinweis (Regelbezeichnung)"] = info.item()
    pruefliste.iloc[index] = row.values
```
